```python
# Import necessary libraries
import numpy as np
import matplotlib.pyplot as plt
from tensorflow.keras.datasets import cifar10
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Input, Dense, Dropout, Flatten
from tensorflow.keras.utils import to_categorical
from tensorflow.keras.optimizers import Adam
from tensorflow.keras.callbacks import EarlyStopping
from sklearn.model_selection import train_test_split
from sklearn.metrics import confusion_matrix
import seaborn as sns  # For better visualization of the confusion matrix

# Load the CIFAR-10 dataset
(x_train, y_train), (x_test, y_test) = cifar10.load_data()

# Preprocessing: Normalize the input data between 0 and 1
x_train = x_train.astype('float32') / 255.0
x_test  = x_test.astype('float32') / 255.0

# Convert the labels to one-hot encoding
num_classes = len(np.unique(y_train))
y_train_cat = to_categorical(y_train, num_classes)
y_test_cat  = to_categorical(y_test, num_classes)

# Split data into training and validation sets
x_train_sub, x_val, y_train_sub, y_val = train_test_split(
    x_train, y_train_cat, test_size=0.2, random_state=42
)

# Define Model Architectures

# Model 1: Baseline model with three Dense layers
def create_model_1(input_shape, num_classes):
    model = Sequential()
    model.add(Flatten(input_shape=input_shape))
    model.add(Dense(256, activation='relu'))
    model.add(Dense(128, activation='relu'))
    model.add(Dense(64, activation='relu'))
    model.add(Dense(num_classes, activation='softmax'))
    return model

# Model 2: Extended model with more layers and neurons
def create_model_2(input_shape, num_classes):
    model = Sequential()
    model.add(Flatten(input_shape=input_shape))
    model.add(Dense(512, activation='relu'))
    model.add(Dense(256, activation='relu'))
    model.add(Dense(128, activation='relu'))
    model.add(Dense(64, activation='relu'))
    model.add(Dense(num_classes, activation='softmax'))
    return model

# Model 3: Model 2 with Dropout layers to prevent overfitting
def create_model_3(input_shape, num_classes):
    model = Sequential()
    model.add(Flatten(input_shape=input_shape))
    model.add(Dense(512, activation='relu'))
    model.add(Dropout(0.4))
    model.add(Dense(256, activation='relu'))
    model.add(Dropout(0.3))
    model.add(Dense(128, activation='relu'))
    model.add(Dropout(0.2))
    model.add(Dense(64, activation='relu'))
    model.add(Dropout(0.1))
    model.add(Dense(num_classes, activation='softmax'))
    return model

# Compile parameters
loss = 'categorical_crossentropy'
metrics = ['accuracy']

# EarlyStopping callback to prevent overfitting
early_stopping = EarlyStopping(monitor='val_loss', patience=5, restore_best_weights=True)

# Placeholder dictionaries to store histories, models, and predictions
histories = {}
models = {}
y_preds = {}  # To store predictions for each model

# Function to compile and train a model
def train_model(model, x_train, y_train, x_val, y_val, epochs=50, batch_size=64):
    # Create a new optimizer instance for each model
    optimizer = Adam(learning_rate=0.001)
    model.compile(optimizer=optimizer, loss=loss, metrics=metrics)
    history = model.fit(
        x_train, y_train,
        validation_data=(x_val, y_val),
        epochs=epochs,
        batch_size=batch_size,
        callbacks=[early_stopping],
        verbose=1
    )
    return history

# Function to compute and plot confusion matrix
def plot_confusion_matrix(y_true, y_pred, class_names, model_name):
    cm = confusion_matrix(y_true, y_pred)
    plt.figure(figsize=(10, 8))
    sns.heatmap(cm, annot=True, fmt='d', cmap='Blues',
                xticklabels=class_names, yticklabels=class_names)
    plt.title(f'Confusion Matrix for {model_name}')
    plt.ylabel('Actual Label')
    plt.xlabel('Predicted Label')
    plt.show()

# Define class names
class_names = [
    'Airplane', 'Automobile', 'Bird', 'Cat', 'Deer',
    'Dog', 'Frog', 'Horse', 'Ship', 'Truck'
]

# Train Model 1
print("Training Model 1...")
model1 = create_model_1(x_train.shape[1:], num_classes)
history1 = train_model(model1, x_train_sub, y_train_sub, x_val, y_val)
histories['Model 1'] = history1
models['Model 1'] = model1

# Predict and compute confusion matrix for Model 1
y_pred_probs_1 = model1.predict(x_test)
y_pred_classes_1 = np.argmax(y_pred_probs_1, axis=1)
y_true_classes = y_test.flatten()
y_preds['Model 1'] = y_pred_classes_1

# Plot confusion matrix for Model 1
plot_confusion_matrix(y_true_classes, y_pred_classes_1, class_names, 'Model 1')

# Train Model 2
print("\nTraining Model 2...")
model2 = create_model_2(x_train.shape[1:], num_classes)
history2 = train_model(model2, x_train_sub, y_train_sub, x_val, y_val)
histories['Model 2'] = history2
models['Model 2'] = model2

# Predict and compute confusion matrix for Model 2
y_pred_probs_2 = model2.predict(x_test)
y_pred_classes_2 = np.argmax(y_pred_probs_2, axis=1)
y_preds['Model 2'] = y_pred_classes_2

# Plot confusion matrix for Model 2
plot_confusion_matrix(y_true_classes, y_pred_classes_2, class_names, 'Model 2')

# Train Model 3
print("\nTraining Model 3...")
model3 = create_model_3(x_train.shape[1:], num_classes)
history3 = train_model(model3, x_train_sub, y_train_sub, x_val, y_val)
histories['Model 3'] = history3
models['Model 3'] = model3

# Predict and compute confusion matrix for Model 3
y_pred_probs_3 = model3.predict(x_test)
y_pred_classes_3 = np.argmax(y_pred_probs_3, axis=1)
y_preds['Model 3'] = y_pred_classes_3

# Plot confusion matrix for Model 3
plot_confusion_matrix(y_true_classes, y_pred_classes_3, class_names, 'Model 3')

# Evaluate the models on the test data
test_metrics = {}
for name, model in models.items():
    test_loss, test_acc = model.evaluate(x_test, y_test_cat, verbose=0)
    test_metrics[name] = {'loss': test_loss, 'accuracy': test_acc}
    print(f"{name} - Test Loss: {test_loss:.4f}, Test Accuracy: {test_acc:.4f}")

# Plot training and validation metrics
def plot_metrics(histories):
    plt.figure(figsize=(18, 6))

    # Plot accuracy
    plt.subplot(1, 2, 1)
    for name, history in histories.items():
        plt.plot(history.history['accuracy'], label=f'{name} Train Acc')
        plt.plot(history.history['val_accuracy'], label=f'{name} Val Acc')
    plt.title('Model Accuracy')
    plt.xlabel('Epoch')
    plt.ylabel('Accuracy')
    plt.legend()

    # Plot loss
    plt.subplot(1, 2, 2)
    for name, history in histories.items():
        plt.plot(history.history['loss'], label=f'{name} Train Loss')
        plt.plot(history.history['val_loss'], label=f'{name} Val Loss')
    plt.title('Model Loss')
    plt.xlabel('Epoch')
    plt.ylabel('Loss')
    plt.legend()

    plt.show()

plot_metrics(histories)

# Compare test accuracies
print("\nTest Accuracy Comparison:")
for name, metrics in test_metrics.items():
    print(f"{name} - Test Accuracy: {metrics['accuracy']:.4f}")

# Choose the best model based on test accuracy
best_model_name = max(test_metrics, key=lambda x: test_metrics[x]['accuracy'])
best_model = models[best_model_name]
print(f"\nBest Model: {best_model_name}")

# Use the predictions from the best model
best_y_pred_classes = y_preds[best_model_name]

# Display some test images with their predicted and true labels
num_images = 10
plt.figure(figsize=(15, 5))
for i in range(num_images):
    plt.subplot(2, 5, i+1)
    plt.imshow(x_test[i])
    plt.title(f"True: {class_names[y_true_classes[i]]}\nPred: {class_names[best_y_pred_classes[i]]}")
    plt.axis('off')
plt.tight_layout()
plt.show()
```
